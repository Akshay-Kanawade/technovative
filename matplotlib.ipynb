# Line plot

```python
x=[3,5,7,9,12,45,68,100]
y=[100,300,900,400,600,200,700,500]
y1=[1000,700,300,500,1200,100,600,900]
plt.style.use('dark_background')
#plt.xkcd()
plt.plot(x,y,color='b',marker='o')
plt.plot(x,y1,color='r',marker='.',linestyle='--')
plt.legend(['F','S'])
plt.grid()
plt.tight_layout()
plt.xlabel("x-point")
plt.ylabel("y-point")
plt.title('Simple line plot')

plt.savefig("temp.jpg")
plt.show()
```

```python
x=[3,13,29,40,65,78,90,100]
y=[100,700,900,500,1200,200,700,1200]
y1=[500,300,300,400,600,100,600,500]

plt.style.use('dark_background')
plt.bar(x,y,color='r',label='f')
plt.bar(x,y1,color='b',label="s")
plt.xlabel('x-point')
plt.ylabel('y-point')
plt.title('bar ')
plt.legend()
plt.show()
```
